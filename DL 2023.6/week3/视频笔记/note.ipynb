## 3.1 Convolutional Neural Network (CNN，卷积神经网络)

图像转化为三维的Tensor：长、宽和通道（Channel，如RGB）

一般图片的像素比较多，采用Fully Connected Network会导致层的超参数很多

如何简化：

- AI通过提取图片的特征来识别，特征可能是图片里的一小部分  
  建立一个感受野（Receptive Field），对于一个Neruon，只计算其感受野里面的数据  
image.png

  感受野可以任意设计

  一个经典的款式：
  - 看所有的channel，只考虑长和宽，即kernel_size（如上图为3x3）
  - stride：感受野向右移动的步数
  - padding：边缘补充，以让感受野能够照顾到边边角角的地方



- 相同的特征会出现在图片的不同区域（如鸟嘴可能在图片的左上角或中间出现）  
  parameter sharing：使得两个neuron的权重和bias一样  
  相同感受野区域的neuron不会共享参数  
  一个经典的款式：
  - Filter：一个neuron公用的参数

Convolutional Layer：Receptive Field + parameter sharing = Filter扫过一次

Feature Map：一个Filter通过输入计算出的结果

后续层的感受野在输入图片上对应着更大的区域，因此感受野很小不会影响对某些较大feature识别的丢失

| Neuron Version Story | Filter Version Story |
|  ----  | ----  |
| Each neuron only considers a receptive field. | There are a set of filters detecting small patterns. |
 | The neurons with different receptive fields share the parameters. |  Each filter convolves over the input image. |

- Pooling（池化）：subsampling，不是layer，是一个operator
  - max pooling：在感受野里面选择最大数值
  - mean pooling：...选平均数值
  - Convolution 和 Pooling 往往交替使用，但对于需要侦测细小特征的模型，pooling会导致较差的效果

- Flatten：把矩阵拉直（把所有数值变为一维的向量）
